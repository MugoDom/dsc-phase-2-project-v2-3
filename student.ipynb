### Objective 2

The second objective was to check factors of housing and their relationship to grading so as to see which ones will increase the grade of a house. This can be shown by a positive corelation.

```python
housing_df.corr()['grade']
```

```python
from sklearn.model_selection import train_test_split
#set up heatmap
fig, ax = plt.subplots(figsize=(17, 12))
corr = housing_df.corr() #use for correlation
sns.heatmap(corr, cmap = 'YlOrRd', annot = True)
# Add title
ax.set_title("Target Variable Correlation Heatmap");

plt.show()
```

We see that sqft_living, sqft_above, price and sqft_living15 have a relatively strong positive corelation. We will now make a new dataframe with just these 5 columns to make it easier to work with.

A correlation of 0.7 to 1.0 (-0.7 to -1) proposes a strong positive (negative) linear relationship while 0.5 to 0.6 (-0.5 to -0.6) is to be considered moderate.

```python
object_2_df = housing_df[['grade','sqft_living', 'sqft_above', 'price', 'sqft_living15']]
object_2_df
```

```python
fig, ax = plt.subplots(figsize=(17, 12))
corr = object_2_df.corr() #use for correlation
sns.heatmap(corr, cmap = 'YlOrRd', annot = True)
# Add title
ax.set_title("Target Variable Correlation Heatmap");

plt.show()
```

```python
#quickly scan for linearity
sns.pairplot(object_2_df, diag_kind = 'hist', corner = True)
```

```python
import statsmodels.api as sm
# Prepare y and X for modeling
y = object_2_df['grade']
X = object_2_df[['sqft_living', 'sqft_above', 'price', 'sqft_living15']]

# use statsmodels
model = sm.OLS(y, sm.add_constant(X))
# fit model
results = model.fit()
# display results
results.summary()
```

The model is statistically significant overall, with an F-statistic p-value well below 0.05
The model explains about 68% of the variance in grade. 
All the independent variables have their p-values below 0.05 and are therefore statistically significant

```python
#Model with Many Features Visualization
fig = plt.figure(figsize=(15,8))
sm.graphics.plot_partregress_grid(
    results,
    exog_idx=list(X.columns.values),
    grid=(2,4),
    fig=fig)
plt.show()
```
